#### 1 Menggunakan data (WBC) dan alogirtma yang sama (DT) pada jobsheet js11_02, silahkan implemnetasikan ulang teknik *exhausive search* dengan menggunakan ***random search***. Scikit-learn sudah menyediakan fungsi ini.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/wbc.csv')

# slice kolom terakhir --> Unnamed column
df = df.iloc[:,:-2]

df.head()
```

```python
X = df.iloc[:, 3:]
y = df['diagnosis']

X.head()
```

```python
from sklearn.model_selection import RepeatedKFold

# RepeatedKFold
cv = RepeatedKFold(n_splits=3, n_repeats=2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV

# inisialisasi model
dt = DecisionTreeClassifier()


params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': list(range(5,11))
}

# inisiasi grid berdasarkan nilai repeated k-fold dan hyperparameter
grid = RandomizedSearchCV(dt, params, cv=cv)



# evaluasi score
grid = grid.fit(X, y)
score = grid.score(X,y)
error_score='raise'

print(f'Evaluation result: {score}')
print(f'hyperparameter configuration: {grid.best_params_}')
```

#### 2 Menggunakan data yang sama, lakukan tunning dengan menggunakan algoritma SVM. Hyperparamter yang akan ditunning adalah,
    - kernel (linier, polinomial, rbf)
    - degree (akan berdampak jika polinomial)
    - koefisien C (regularization)

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/wbc.csv')

# slice kolom terakhir --> Unnamed column
df = df.iloc[:,:-2]

df.head()
```

```python
X = df.iloc[:, 2:]
y = df['diagnosis']

X.head()
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=32)
```

```python
print; print(X_train.shape)
print; print(y_train.shape)
```

```python
svm = SVC()
svm.fit(X_train,y_train)
svm.score(X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = { 'C':[0.1,1,100,1000],'kernel':['rbf','poly','linear'],'degree':[1,2,3,4,5,6]}
grid = GridSearchCV(SVC(),param_grid)
grid.fit(X_train,y_train)
```

```python
print(grid.best_params_)
print(grid.score(X_test,y_test))
```
